```python
filelist = glob.glob('*')
filelist
strange_file = ['16_쌀밥', '17_계란후라이', '18_계란말이'] # 이미지 파일 이름 변경 필요
for rm_file in strange_file:
    filelist.remove(rm_file)

for file_name in filelist:
    img_list = glob.glob(file_name + '/*')
    for i, img_name in enumerate(img_list):
        img_list[i] = img_name.lstrip(file_name + '/')
        
    for img_name in img_list:
        shutil.move(file_name + '/' + img_name,img_name)
```

```python
# strange_file (이미지 파일 이름 바꿀 것들)
cnt = 150
for file_name in strange_file:
    cnt += 1
    img_list = glob.glob(file_name + '/*')
    old_img_list = img_list.copy()
    for i, img_name in enumerate(img_list):
        for_new_name = img_name.lstrip(file_name + '/').split('_')
        img_list[i] = for_new_name[0] + '_' + str(cnt) + '_' + for_new_name[1]

    for num, img_name in enumerate(img_list):
        shutil.move(old_img_list[num], img_name)
```
