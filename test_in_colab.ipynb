```python
os.mkdir('../trans잡곡밥')
```

```python
# 바꿀 숫자 
trans_num = 0
with open(aa, 'r') as f:
    txtline = ''
    lines = f.readlines()
    for line in lines:
        # 첫번째 공백 찾기
        first_blank = line.find(' ')

        # 공백앞의 문자
        pre_num = line[0:first_blank]

        # 공백앞의 문자 제거
        rm_pre_num = line.lstrip(pre_num)

        # 바꿀 숫자 삽입
        after_line = str(trans_num) + rm_pre_num

        # data
        txtline += after_line
    with open('../trans잡곡밥/' + aa, 'w') as tf:
        tf.write(txtline)
```
